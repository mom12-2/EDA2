Goal: Find out what factors drive higher tips at a restaurant.
Num vs. Num - Scatter: Create a sns.scatterplot() with total_bill on the X-axis and tip on the Y-axis to prove that higher bills equal higher tips.

```python
import matplotlib.pyplot as plt
import seaborn as sns
df= sns.load_dataset('tips') 
df
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

The scatter plot shows a positive relationship between total bill and tip, meaning higher bills usually receive higher tips. However, the pattern is not perfect because some values vary and include outliers. Most data points are concentrated in the mid-range bill amounts.

    ក្រាប Scatter Plot បង្ហាញទំនាក់ទំនងវិជ្ជមានរវាង total bill និង tip មានន័យថា ពេល bill កើនឡើង tip ក៏មានទំនោរកើនឡើងដែរ។ ប៉ុន្តែមិនមែនត្រឹមត្រូវគ្រប់ករណីទេ ព្រោះមានភាពខុសគ្នា និងចំណុចខុសធម្មតាខ្លះ។ ទិន្នន័យភាគច្រើនស្ថិតនៅចន្លោះមធ្យមនៃ bill។

Num vs. Num - Line: Create a mock timeline of total_bill values using sns.lineplot() to practice connecting data points.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
mock_df = pd.DataFrame({
    'month': months,
    'total_bill': df['total_bill'][:12].values
})

sns.lineplot(x='month', y='total_bill', data=mock_df)

plt.show()
```

The line plot shows total_bill values across months (Jan–Dec) using the first 12 data points. It is used to visualize changes over time, showing increases or decreases in values, but it is not real monthly data since the months are only labels.

    ក្រាប Line Plot បង្ហាញតម្លៃ total_bill តាមខែ (Jan–Dec) ដោយប្រើតម្លៃ 12 ដំបូងពី dataset។ វាប្រើសម្រាប់បង្ហាញការប្រែប្រួលនៃទិន្នន័យតាមលំដាប់ ប៉ុន្តែមិនមែនជាទិន្នន័យពិតតាមខែទេ ព្រោះខែគ្រាន់តែជា label ប៉ុណ្ណោះ។

Num vs. Cat - Box Plots: Use sns.boxplot() with day on the X-axis and total_bill on the Y-axis to find the median bill and outliers for each day of the week.

```python
sns.boxplot(x='day', y='total_bill', data=df)

plt.show()
```

The box plot shows how total bill values are distributed across different days. It displays the median, middle range, and outliers for each day, helping compare spending patterns and variation between days.

    ក្រាប Box Plot បង្ហាញការចែកចាយតម្លៃ total_bill តាមថ្ងៃនីមួយៗ។ វាបង្ហាញមេឌៀន, ចន្លោះកណ្ដាល និងចំណុចខុសធម្មតា (outliers) ដើម្បីជួយប្រៀបធៀបទម្លាប់ចំណាយ និងភាពខុសគ្នារវាងថ្ងៃនីមួយៗ។

Num vs. Cat - Violin Plots: Use sns.violinplot() comparing time (Lunch vs. Dinner) and tip to see where the density of generous tippers lies.

```python
sns.violinplot(x='time', y='tip', data=df)

plt.show()
```

The violin plot shows the distribution of tip amounts for Lunch and Dinner. Wider sections represent areas with more data, while narrower sections represent fewer data points. It helps compare the spread and concentration of tips between the two meal times.

Violin Plot បង្ហាញការចែកចាយនៃចំនួន tip រវាង Lunch និង Dinner។ ផ្នែកទូលាយមានន័យថាមានទិន្នន័យច្រើន ខណៈផ្នែកស្តើងមានទិន្នន័យតិច។ វាជួយប្រៀបធៀបការប្រែប្រួល និងការប្រមូលផ្តុំនៃ tip ក្នុងពេលអាហារទាំងពីរ។

Num vs. Cat - Bar Plots: Use sns.barplot() to show the average tip amount grouped by size (party size), letting Seaborn calculate the mean automatically.

```python
sns.barplot(x='size', y='tip', data=df)

plt.show()
```

The bar plot shows the average tip amount for different table sizes. It helps compare how the number of people at a table affects tipping behavior. Larger groups generally leave higher average tips than smaller groups.

Bar plot នេះបង្ហាញពីចំនួន tip មធ្យម សម្រាប់ទំហំតុខុសៗគ្នា។ វាជួយប្រៀបធៀបថាចំនួនមនុស្សនៅតុមានឥទ្ធិពលលើការទុក tip យ៉ាងដូចម្តេច។ ជាទូទៅ ក្រុមមនុស្សធំៗ ទុក tip មធ្យមច្រើនជាងក្រុមតូចៗ។
